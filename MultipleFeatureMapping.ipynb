```python
def feature_mapping(df1, df2):
    name=df1.columns[0]
    feature=df1.columns[1:]
    feature_len=len(feature)
    sub_feat_dict=[dict(zip(df1[name], df1[i])) for i in feature]
    
    for i, j in zip(feature, range(feature_len)):
        df2[i]=df2[name].map(sub_feat_dict[j])
```
