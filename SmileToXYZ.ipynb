```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
smile='CC1=CC(C)=CC(C)=C1'
```

```python
mol=Chem.MolFromSmiles(smile) #Creating rdkit-mol object
```

```python
hmol=Chem.AddHs(mol) # adding H to the system
```

```python
hmol #just checking it 
```

```python
#For bigger and complex system this step fails sometime
AllChem.EmbedMolecule(hmol, AllChem.ETKDG()) #Embedd the molecule
AllChem.UFFOptimizeMolecule(hmol, 1000) #optimization of the molecule using UFF
```

```python
hmol #check it again after genereation of conformation and optimization
```

```python
Chem.MolToXYZFile(hmol, 'molecule1.xyz') #writing the coordinates in a file
```
